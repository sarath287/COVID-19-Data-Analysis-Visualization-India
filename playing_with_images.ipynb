```python
import cv2 # computer vision library to work with images
import numpy as np
import matplotlib.pyplot as plt

def display_img(name,img):
  plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
  # plt.imshow(img)
  plt.title(name)
  # plt.axis('off')
  plt.show()
display_img("MY",cv2.imread("my child photo.jpg"))
```
